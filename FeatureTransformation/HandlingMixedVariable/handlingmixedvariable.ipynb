```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../../EDA/data/titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot(kind='bar')
fig.set_title('number of people')
```

```python
df['numerical_number'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
```

```python
df['number_categroical'] = np.where(df['numerical_number'].isnull(), df['number'],  np.nan)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\\d+)')
df['cabin_cat'] = df['Cabin'].str[0]
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
df['ticket_num'] = df['Ticket'].apply(lambda x: x.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce', downcast='integer')
```

```python
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])
```
