OHE using sklearn

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=42
)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32)
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel', 'owner']])
X_test_new = ohe.transform(X_test[['fuel', 'owner']])
```

ONE HOT ENCODING for large number of categorical datas
We split it into 2, frequent and none frequent


```python
counts = df['brand'].value_counts()
counts
```

```python
threshold = 100
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'other'), dtype=np.int32).sample(5)
```
