```python
#date is in object format, we need to convert it to datetime format
date.info()
```

```python
#converting date column to datetime format using pandas to_datetime function
date['date'] = pd.to_datetime(date['date'])
date.info()
```

```python
#Year
date['date_year'] = date['date'].dt.year
date.sample(5)
```

```python
#Month and day

date['date_month'] = date['date'].dt.month
date['date_day'] = date['date'].dt.day
date.sample(5)
```

```python
#Month name
date['month_name'] = date['date'].dt.month_name()
date.sample(5)
```

```python
#day name

date['day_name'] = date['date'].dt.day_name()
date.sample(5)
```

```python
#day of the week
date['day_of_week'] = date['date'].dt.dayofweek
date.sample(5)
```

```python
#is weekend or not (1: weekend, 0: weekday)
date['is_weekend'] = np.where(date['day_name'].isin(['Saturday']), 1, 0)
date.sample(5)
```

```python
date['quarter'] = date['date'].dt.quarter
date.sample(5)
```

```python
date['semester'] = np.where(date['quarter'].isin([1, 2]), 1, 2)
date
```
