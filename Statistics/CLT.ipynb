```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample_size = 60
total_samples = 1000


sample = np.random.uniform(1, 100, size=60)
sample_mean = np.mean(sample)
sample_mean
```

```python
sample_means = []
for i in range(total_samples):
    sample = np.random.randint(1, 100, size=sample_size)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)
```

```python

sns.histplot(sample_means, color='b', fill=True, kde=True)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

The probability distribution of the sample mean tends to follow a normal distribution, regardless of the original population distribution, provided the sample size is sufficiently large. This result is explained by the Central Limit Theorem (CLT).

According to the Central Limit Theorem, if random samples of size ùëõ are drawn from a population with mean ùúá and variance (ùúé^2), then the sampling distribution of the sample mean (ùëã') approaches a normal distribution as ùëõ becomes large. The mean of this sampling distribution is ùúá, and its standard deviation (called the standard error) is ùúé/ùëõ.

Therefore, even if the population distribution is not normal, we can use the normal distribution to make statistical inferences about the population mean when the sample size is large. This allows us to construct confidence intervals and perform hypothesis testing about the population parameter.

```python
clt_mean = np.mean(sample_means)
clt_std = np.std(sample_means)
clt_mean, clt_std
```

```python
sample_mean1 = []
for i in range(total_samples):
    sample1 = np.random.poisson(lam=5, size=sample_size) # Poisson distribution with lambda=5
    sample_mean1.append(np.mean(sample1))

    
```

```python
sns.histplot(sample_mean1, color='g', fill=True, kde=True)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
sample_mean2 = []
for i in range(total_samples):
    sample2 = np.random.exponential(scale=1.0, size=sample_size) # Exponential distribution with scale=1.0
    sample_mean2.append(np.mean(sample2))
sns.histplot(sample_mean2, color='r', fill=True, kde=True)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```
