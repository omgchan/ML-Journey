```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
df['sepal_length'].plot(kind='kde')
```

```python
temp = np.sort(df['sepal_length'])
```

```python
y_quant = []
for i in range(len(df)):
    y_quant.append(np.percentile(temp, (i + 1) * 100 / len(df)))
```

```python
sample = np.random.normal(loc=0, scale=1, size=150)
```

```python
x_quant = []
for i in range(len(sample)):
    x_quant.append(np.percentile(sample, (i + 1) * 100 / len(sample)))
```

```python
sns.scatterplot(x=x_quant, y=y_quant)

```

```python
import statsmodels.api as sm

fig = sm.qqplot(df['sepal_length'], line='45', fit=True )

plt.title('QQ Plot of Sepal Length')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```
