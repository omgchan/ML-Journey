```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import shapiro, t, levene
```

```python
age = pd.read_csv('../../EDA/data/train.csv', usecols=['Age'])
age = age.dropna()
age
```

```python
sample_age = age.sample(25).values.flatten()
sample_age
```

```python
# H0: μ = 30 (The mean age of the population is 30 years)
# H1: μ > 30 (The mean age of the population is greater than 30 years)
# Step 1: Calculate the sample mean, standard deviation, and size
# Level of significance (α) = 0.05

sample_mean = np.mean(sample_age)
sample_size = len(sample_age)
sample_std = np.std(sample_age, ddof=1)  # Sample standard deviation
```

```python
shapiro_test = shapiro(sample_age)
print(f"Shapiro-Wilk test for normality: {shapiro_test}")

# shaipro_test >  0.05 means the data is normally distributed, we can proceed with the t-test
```

```python
t_statistic, p_value = stats.ttest_1samp(sample_age, popmean=30, alternative='greater')
print(f"One-sample t-test statistic: {t_statistic}, p-value: {p_value}")
```

```python
if p_value < 0.05:
    print("Reject the null hypothesis (H0). The mean age of the population is greater than 30 years.")
else:
    print("Fail to reject the null hypothesis (H0). There is no evidence to suggest that the mean age of the population is greater than 30 years.")
```
