```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv('../../EDA/data/train.csv', usecols=['Age', 'Sex'])
df
```

```python
df = df.dropna()
df.sample(5)
```

```python
pop = df['Age'].tolist()
pop_mean = np.mean(pop)
pop_mean
```

```python
df.Age.describe()
```

```python
pop_male = df[df['Sex'] == 'male']['Age'].tolist()
pop_female = df[df['Sex'] == 'female']['Age'].tolist()
```

```python
pd.Series(pop_male).isna().sum()
pd.Series(pop_female).isna().sum()
```

```python
# H0: the mean age of male = female (Null hypothesis: there is no significant difference between the mean ages of male and female)
# H1: the mean age of male > female (Altenative hypothesis is one tailed)

sample_male = np.random.choice(pop_male, size=25, replace=False)
sample_female = np.random.choice(pop_female, size=25, replace=False)
alpha = 0.05
```

```python
from scipy.stats import shapiro, levene, ttest_ind

shapiro_result_male = shapiro(sample_male)
shapiro_result_female = shapiro(sample_female)
print("Shapiro-Wilk test for male sample:", shapiro_result_male)
print("Shapiro-Wilk test for female sample:", shapiro_result_female)
```

shapiro values of male and female > alpha (0.05): so the sample is normal

```python
levene_test = levene(sample_male, sample_female)
print("Levene's test for equality of variances:", levene_test)
```

p_value > 0.05(alpha): so the two samples have no significance differnce in varience

```python
t_stats, p_value = ttest_ind(sample_male, sample_female)
print("T-test statistic:", t_stats)
print("P-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: there is a significant difference in mean ages between male and female")
else:
    print("Fail to reject the null hypothesis: there is no significant difference in mean ages between male and female")
```
