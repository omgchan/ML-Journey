```python
import pandas as pd
import random
```

```python
#undestanding probability distributions (Probability Mass Function *PMF* by simulating dice rolls)
L = []
for i in range(10000):
  a = random.randint(1,6)
  b = random.randint(1,6)
  L.append(a+b)
```

```python
len(L)
```

```python
L[:5]
```

```python
#calulating s which is the prbabilities of the possible outcomes (2 to 12) of rolling two dice
a = pd.Series(L).value_counts()  #counts of each outcome
b = pd.Series(L).value_counts().sum() #total number of outcomes
s = (a/b).sort_index() #probabilities of each outcome sorted by index (outcome value)
```

```python
s #The PMF showing the probabilities of each possible outcome when rolling two dice
```

```python
s.plot(kind='bar', title='Probability Mass Function (PMF) of Two Dice Rolls', xlabel='Sum of Dice', ylabel='Probability')
```

```python
# Demonstrating the CDF of the PMF
import numpy as np
```

```python
c = np.cumsum(s) #Cumulative sum to get CDF
```

```python
c.plot(kind='bar', title='Cumulative Distribution Function (CDF) of Two Dice Rolls', xlabel='Sum of Dice', ylabel='Cumulative Probability')
```

**PROBABILITY DESNITY FUNCTION**

PDF cannot be directly calculated like pmf does because the y-axis in the pdf represents **Probability Desnity**. In order to plot the graph of the data, we use density estimation by calulcating respective probabilty density.

Density Estimation is futher divided into 2 parts: **Parametric and Non-Parametric**. Simply, *paramteric is where the graph follows specific parametric family of the Probability distrubutin, while non-parametric doesn't.*

```python
#Showcasing the PDF
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
```

```python
sample = normal(loc=50, scale=5, size=1000)
```

```python
plt.hist(sample, bins=10, alpha=0.6, color='g')
```

```python
sample_mean = sample.mean()
sample_mean
```

```python
sample_std = sample.std()
sample_std
```

```python
from scipy.stats import norm
dist = norm(sample_mean, sample_std)
```

```python
values = np.linspace(sample.min(), sample.max(), 100)
```

```python
sample.max()
```

```python
probabilty_densities = [dist.pdf(value) for value in values ]
```

```python
plt.hist(sample, bins=10, density=True)
plt.plot(values, probabilty_densities)
```

```python
#also you could use seaborn to directly calculate and plot the PDF
import seaborn as sns
```

```python
sns.displot(sample, kde=True)
```

**KERNEL DESNITY ESTIMATION**
KDE is used when the data follows non-parametric pattern.

```python
#creating samples from a mixture of two normal distributions

sample1 = normal(loc=20, scale=5, size=300 )
sample2 = normal(loc=40, scale=5, size=700 )

sample = np.hstack((sample1, sample2))
```

```python
sample
```

```python
plt.hist(sample, bins=50)
```

```python
#Fitting a KDE to the mixture distribution
from sklearn.neighbors import KernelDensity
```

```python
#**bandwidth** = standard deviation of the kernel function. 
# Its value influences the smoothness of the resulting density estimate.
#  A smaller bandwidth leads to a more sensitive estimate that captures
#  more details (but may also capture noise),
#  while a larger bandwidth produces a smoother
#  estimate that may overlook finer structures in the data.

model = KernelDensity(bandwidth=3, kernel='gaussian') 

sample = sample.reshape(len(sample), 1) #reshaping to 2D array as required by sklearn
model.fit(sample)
```

```python
values = np.linspace(sample.min(), sample.max(), 100) #creating values over which to evaluate the KDE
values = values.reshape(len(values),1)
```

```python
probabilities = model.score_samples(values)
probabilities = np.exp(probabilities) # converting log densities to densities (ln(e^x) = x)
```

```python
plt.hist(sample, bins=50 , density=True)
plt.plot(values[:], probabilities)
plt.show()
```

```python
sns.kdeplot(sample.reshape(len(sample)), bw_adjust=2)
```

**Proability Distributions Summary:**
- Probability distributions describe how the values of a random variable are distributed.
- PMF is used for discrete variables, showing the probability of each outcome.
- CDF represents the cumulative probability up to a certain value.


- PDF is used for continuous variables, estimating the density of data points.
- Unlike PMF, PDF values do not represent direct probabilities but rather densities. 
-So the area under the PDF curve over an interval gives the probability of the variable falling within -that interval.

-Here, PDF can be generated using either *Parametric methods* (assuming a specific distribution) or
*Non-Parametric methods* (like KDE) that do not assume a specific distribution form.
**bandwidth** in KDE controls the smoothness of the estimated density curve.
