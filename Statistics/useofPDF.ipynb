**Trying to understand the use of Probabilty Density Function**
PDFs are specially used in feature selection, trying to understand the impact of each features in the result.
The code below will demonstrate the use of PDF in feature selection

```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
df.sample(5)
```

iris is an inbuilt dataset in the seaborn.
Here
We can see the result of *df.sample(5)* shows the 5 samples of data, where 4 features help us to determine the result *species*

```python
sns.kdeplot(data=df, x="sepal_length", hue="species", fill=True)
```

**PDF of sepal_lengths** of each flower is overlaped, so it is diffcult to determine the impact of the sepal_length.
If sepal_length = 6, then the flower could be anyone of them here.  

```python
sns.kdeplot(data=df, x="sepal_width", hue="species", fill=True)
```

It's the same intuition here as the above. **PDF of sepal_width** of each flower is overlaped, so it is diffcult to determine the impact of the sepal_length.
If sepal_length = 6, then the flower could be anyone of them here.  

```python
sns.kdeplot(data=df, x="petal_length", hue="species", fill=True)
```

### Effect of Petal Length on Class Prediction

The probability density functions (PDFs) indicate that **petal length** is a highly informative feature for class prediction. The *Setosa* class is linearly separable, with its distribution confined to the range **0.5 < petal_length < 2.5**, resulting in negligible overlap with other classes.

The *Versicolor* and *Virginica* classes occupy higher petal length ranges, primarily **2.55.5** and **> 5.5**, respectively. Although their PDFs exhibit partial overlapintroducing ambiguity near the decision boundarythe majority of their probability mass remains separable.

Overall, petal length provides strong discriminative power on its own; however, accurate separation between *Versicolor* and *Virginica* can be further improved by incorporating additional features.
