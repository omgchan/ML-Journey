```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('../EDA/data/deaths.csv')
```

```python
df1
```

```python
df1.info()
```

```python
df = df1[df1['Year'] == 2000]
df
```

```python
df['total_deaths'] = df.iloc[:,4:].sum(axis=1, skipna=True)
```

```python
total_deaths = np.array(df['total_deaths'].to_list())
total_deaths
```

```python
samples_means = []
total_samples = 1000
sample_size = 50

for _ in range(total_samples):
    sample = np.random.choice(total_deaths, size=sample_size, replace=True)
    samples_means.append(np.mean(sample))

```

```python
sns.histplot(samples_means, kde=True, alpha=0.5, color='blue')
plt.title("Distribution of the sample means")
```

Since deaths cannot be negative, the distribution of total deaths is bounded below by zero. This creates a concentration of values near smaller numbers, while a few extreme high values extend the right tail. As a result, the population distribution is slightly right-skewed. Despite this skewness, the sampling distribution of the mean becomes approximately normal for sufficiently large sample sizes, illustrating the Central Limit Theorem in practice.

```python
print(np.mean(total_deaths))
np.mean(samples_means)
```

```python
print((np.std(total_deaths))/np.sqrt(sample_size))
print((np.std(samples_means)))
```

```python
from scipy.stats import skew
skew(total_deaths)
skew(samples_means)
```

```python
sample_mean1 = []
for _ in range(total_samples):
    sample = np.random.choice(total_deaths, size=1000, replace=True)
    sample_mean1.append(np.mean(sample))
```

```python
sns.histplot(sample_mean1, kde=True, alpha=0.5, color='red')
plt.title("Distribution of the sample means with sample size 1000")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

When the sample size was increased from 50 to 1000, the sampling distribution of the mean became more symmetric and closer to a normal distribution, even though the underlying population and its parameters remained the same. This demonstrates the Central Limit Theorem, which states that the distribution of the sample mean approaches normality as the sample size increases, regardless of the shape of the original population.

```python
print(np.mean(sample_mean1))
print((np.std(sample_mean1)))
```

```python
print((np.mean(total_deaths)))
print((np.std(total_deaths))/np.sqrt(1000))
```

### Confidence Interval for the Mean

A confidence interval (CI) gives a range of values within which we are fairly certain the true population mean lies.

1. **Significance Level (α):**  
   For a 95% confidence interval, α = 0.05

2. **Critical Value (Z):**  
   For a two-tailed 95% CI, Z = 1.96

3. **Standard Error (SE):**  
   SE = population standard deviation / sqrt(sample size)

4. **Confidence Interval Formula:**  
   CI = sample mean ± (Z × SE)

5. **Interpretation:**  
   - We are 95% confident that the true population mean lies within this interval.  
   - Increasing the sample size decreases SE, making the confidence interval narrower and the estimate more precise.


```python
se = (np.std(sample_mean1)/np.sqrt(1000))

lower_bound = np.mean(sample_mean1) - 1.96 * se 
upper_bound = np.mean(sample_mean1) + 1.96 * se
print(f"95% Confidence Interval: [{lower_bound} --- {upper_bound}]")
```

```python
print(np.mean(total_deaths))
```

Conclusion: The simulation confirms that the sampling distribution of the sample mean approaches a normal distribution, even if the underlying population is skewed. The mean of the sampling distribution approximates the population mean, while the variance of the sample mean is equal to the population variance divided by the sample size. This illustrates the Central Limit Theorem in practice.
