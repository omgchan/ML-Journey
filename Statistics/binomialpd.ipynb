```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10
p = 0.5
size = 1000

binomialpmf = np.random.binomial(n, p, size)
binomialpmf
```

```python
plt.hist(binomialpmf, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.6, color='g')
plt.title('Binomial Distribution PMF')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.grid()
```

```python

binomialpmf2 = np.random.binomial(n, 0.5, 1_000_000)
plt.hist(binomialpmf2, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.6, color='b')
plt.title('Binomial Distribution PMF')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.grid()
```
