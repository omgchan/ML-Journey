```python
dfs = []

for i in range(1,10):
    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key=ea5ba55c9738b19ac8282c924b149303&language=en-US&page={i}"
    response = requests.get(url)
    temp_df = pd.DataFrame(response.json()['results'])[
        ['id', 'title', 'release_date', 'overview', 'vote_average', 'vote_count']
    ]
    dfs.append(temp_df)

df = pd.concat(dfs, ignore_index=True)
```
