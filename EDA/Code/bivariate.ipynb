```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
titanic = pd.read_csv('train.csv')
flight = sns.load_dataset("flights")
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
```

```python
titanic.head()
```

```python
sns.barplot(data=titanic, x='Pclass', y='Age', hue='Sex')
```

```python
sns.displot(titanic[titanic['Age'].notnull()], x='Age', kde=True, hue='Sex', col='Pclass', bins=30)
```

```python
sns.kdeplot(
    data=titanic[titanic['Survived'] == 1],
    x='Age',
    color='green',
    label='Survived'
)

sns.kdeplot(
    data=titanic[titanic['Survived'] == 0],
    x='Age',
    color='red',
    label='Not Survived'
)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```
