```python
import numpy as np
import matplotlib.pyplot as plt

def linear_least_squares(x, y):
    n = len(x)
    
    # Calculating the required sums
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_xy = np.sum(x * y)
    sum_x_squared = np.sum(x**2)
    
    # Applying the formulas for m and c
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
    c = (sum_y - m * sum_x) / n
    
    return m, c

# 1. Test Data
x_data = np.array([1, 2, 3, 4, 5, 6, 7])
y_data = np.array([1.5, 3.8, 6.7, 9.0, 11.2, 13.6, 16.0])

# 2. Compute Coefficients
slope, intercept = linear_least_squares(x_data, y_data)

# 3. Generate Prediction Line
y_pred = slope * x_data + intercept

# Output Results
print(f"Regression Equation: y = {slope:.4f}x + {intercept:.4f}")

# 4. Visualization
plt.scatter(x_data, y_data, color='red', label='Original Data')
plt.plot(x_data, y_pred, color='blue', label='Best Fit Line')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.title('Linear Least Squares Regression')
plt.legend()
plt.grid(True)
plt.show()
```
