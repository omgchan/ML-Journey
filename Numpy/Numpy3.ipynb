```python
import numpy as np
import random
```

```python
a = np.random.randint(1, 100, 10 )
print(a)
```

np.sort(a) : creates a sorted array

```python
np.sort(a)
```

```python
b = np.random.randint(1, 100, 10 ).reshape((2,5))
b
```

```python
print(np.sort(b, axis=0) ) # sort by column
print(np.sort(b, axis=1) ) # sort by row
```

```python
print(np.append(a, 100)) # append a single value to the end of a
print(np.append(a, np.ones(6))) #append an array of ones to the end of a
print(np.append(b, np.ones((2,5)), axis=0)) # append 2 rows of ones to b
print(np.append(b, np.ones((2,5)), axis=1)) # append 2 columns of ones to b
```

```python
np.concatenate((a, np.ones(6))) # concatenate a and an array of ones
np.concatenate((b, np.ones((2,5))), axis=0) # concatenate b and 2 rows of ones
np.concatenate((b, np.ones((2,5))), axis=1) # concatenate b     
```
