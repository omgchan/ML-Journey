```python
#speed
import time

a = [ i for i in range(1_000_000) ]
b = [ i for i in range(1_000_000, 2_000_000) ]

c = []

start = time.time()

for i in range(len(a)):
    c.append(a[i] + b[i])

print("Time taken for list: ", time.time() - start)
```

```python
x = np.arange(1_000_000)
y = np.arange(1_000_000, 2_000_000)
start = time.time()
z = x + y
print("Time taken for numpy: ", time.time() - start)
```

```python
#fancy indexing
a = np.arange(24).reshape(4, 6)
a
```

```python
# You can access specific rows and columns using fancy indexing
print(a[0]) # first row
print(a[:, 0]) # first column
```

```python
# or you could access specific rows and columns which are not in order
print(a[[0, 2]]) # first and third row
print(a[:, [0, 2]]) # first and third column
```

### BROADCASTING

Broadcasting is the process of making the dimensions and the shape of the arrays equal. Here, the 1 will be added to the head of the smaller array to make of the same dimension and stretch that 1 to match the shape of the array

```python
a = np.arange(12).reshape(3,4)
b = np.random.randint(0, 100, 1)
```

```python
print(a)
print(b)

print(a + b) 
# broadcasting in action; Eventhough a and b have different shapes, numpy broadcasts b to match the shape of a
```

```python
# Creation of Mathematical functions

x = np.arange(10)

def sigmoid(x):
    return (1/(1 + np.exp(-x)))

sigmoid(x)
```

```python
# or MSE
y = np.random.randint(0, 2, 10)

py = np.random.uniform(0.01, 0.99, 10)


def mse(actual, pred):
    return np.mean((actual - pred) ** 2)

mse(y, py)
```
