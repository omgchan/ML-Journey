###INDEXING AND SLICING

```python
a1= np.arange(10)
a2 = np.arange(12).reshape((3,4))
a3 = np.arange(8).reshape((2,2,2))
```

```python
print(a1, "\n")
print(a2, "\n")
print(a3, "\n")

```

```python
#indexing: indexing in 1D array is same as indexing in python
print (a1[1])
print (a1[-1])

```

```python
#For 2D array, we use the index of element by their rows and columns (aij)
a2[1,2] #accssing 6 which is in the 2nd row(index 1) and 3rd col(index 2)
```

```python
#For 3D arrays, first we identify which matrix has the element and use the same concept used in 2D array(matrix)
a3[1,1,0] #accessing 6
```

```python
#slicing
#For 1D array, the concept of slicing is same as python
a1[1:4]
```

```python
a1[1:4:2]
```

```python
a1[:]
```

```python
#For 2D array, the concept is same but have to use in terms of rows and columns
a2[0, :]  #accesing all the elements of the first row

```

```python
a2[:, 2] #accessing all the elements of the 3 column
```

```python
#using silicing you can get multiple datas
a2[:, 1: ]
```

```python
a2[::2, :3:2]
```

```python
a2[0:2,1:]
```

```python
#For 3D,
a3 = np.arange(27).reshape((3,3,3))
a3
```

```python
a3[0, 1, :]
```

```python
a3[2, 1:, 1:]
```

```python
a3[::2, 0, ::2]
```

```python
a3[::2,1, ::2]
```

###ITERATING


```python
print(a1, "\n")
print(a2, "\n")
print(a3, "\n")
```

```python
for i in a1:
  print(i, end="\t")
```

```python
for i in a2:
  for j in i:
    print(j, end="\t")
  print('\n')
```

```python
for i in a3:
  for j in i:
    for k in j:
      print(k, end="\t")
    print('\n')
  print('\n')
```

```python
#or you could use nditer() function to loop through multiple dimension of array

for i in np.nditer(a3):
  print(i)
```

##STACKING


```python
#stacking is like addigng the arrays either vertical or horizontally
s1 = np.random.random(4).reshape((2,2))
s1 = np.floor(s1 * 10)
s2 = np.random.random(4).reshape((2,2))
s2 = np.floor(s2 * 10)
```

```python
x = np.hstack((s1, s2)) #horizontal stacking
x
```

```python
y = np.vstack((s1, s2))
y
```

###SPLITING: OPP OF STACKING


```python
np.hsplit(x, 2 ) #horizontal splitting
```

```python
np.vsplit(y, 1)
```
